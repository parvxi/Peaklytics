```python
from langchain_ollama import OllamaLLM
llm = OllamaLLM(model="llama3.1")
```

```python
market_category = "Technology"
funding_total_usd = 1000000
region = "USA"
city = "New York"
country = "USA"
funding_rounds = 10
seed = 100000
venture = 100000
equity_crowdfunding = 100000
undisclosed = 100000
status = "success"

# Use .format() to replace the placeholders with actual variable values
prompt_success = f'''
The company will most likely {status} based on the following data. As an expert business consultant, please analyze this information and provide detailed insights into the potential reasons for this company. Additionally, offer advice and actionable recommendations on how the company can improve or capitalize on its current situation.

Here are the details for the company:

- Market Category: {market_category}
- Total Funding (USD): {funding_total_usd}
- Region: {region}
- City: {city}
- Number of Funding Rounds: {funding_rounds}
- Seed Funding: {seed}
- Venture Capital Funding: {venture}
- Equity Crowdfunding: {equity_crowdfunding}
- Undisclosed Funding: {undisclosed}
- Country: {country}

Based on this information:

1. Explain what factors contributed to its success, and provide advice on how to maintain or further enhance its growth. What strategies should the company adopt to ensure long-term sustainability and continued success? Suggest potential opportunities or markets that the company could explore to maximize its potential.

2. Provide a few practical examples or case studies of similar companies that succeeded, and explain how they managed to do so.

Please ensure that the insights are clear, actionable, and tailored to the company's specific situation.
'''
```

```python
response = llm.invoke(prompt_success)
print(response)
```
